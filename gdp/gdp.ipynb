# GDP Data
Retrieved from United Nations: https://unstats.un.org/unsd/snaama/Basic

```python
import pandas as pd
import scipy.stats as stats
import numpy as np
```

```python
countries = pd.read_csv('bias/countries.csv')
countries.head()
```

```python
gdp_percapita = pd.read_csv('gdp/GDP_percapita.csv')
gdp = pd.read_csv('gdp/GDP_current.csv')
gdp_constant = pd.read_csv('gdp/GDP_constant2015.csv')
```

```python
gdp_all.
```

```python
gdp_all = gdp_constant.merge(gdp.merge(gdp_percapita, on=['Country/Area', 'Year', 'Unit']), on=['Country/Area', 'Year', 'Unit']).rename({'GDP, at current prices - US Dollars': 'GDP', 'GDP, Per Capita GDP - US Dollars': 'GDP_pc', 'GDP, at constant 2015 prices - US Dollars': 'GDP_constant'}, axis=1)
gdp_all
```

```python
gdp_5_year_avg = gdp_all.groupby('Country/Area').mean().reset_index()
gdp_5_year_avg
```

```python
gdp_5_year_avg[~gdp_5_year_avg['Country/Area'].isin(countries.name)]
```

```python
gdp_5_year_avg[gdp_5_year_avg['Country/Area'].isin(countries.name)][['GDP_pc', 'oecd']].boxplot(by='oecd')
```

# OECD Data
Retrieved from Wikipedia: https://en.wikipedia.org/wiki/OECD (using https://wikitable2csv.ggor.de/)     

```python
oecd_members = pd.read_csv('gdp/oecd_members.csv')
oecd_members.head()
```

```python
oecd_countries = set(oecd_members.Country.values)
gdp_all['oecd'] = gdp_all['Country/Area'].apply(lambda t: t in oecd_countries)
```

```python
replace_dict = {
    'Bolivia (Plurinational State of)': 'Bolivia',
    'United Kingdom of Great Britain and Northern Ireland': 'United Kingdom',
    'Russian Federation': 'Russia',
    'Czechia': 'Czech Republic',
    'Venezuela (Bolivarian Republic of)': 'Venezuela',
    'Republic of Moldova': 'Moldova',
    # '': 'Taiwan',
    'Viet Nam': 'Vietnam', 
    'Republic of Moldova': 'Moldova', 
    #'Democratic Republic of the Congo': 'Republic of the Congo',
    'China (mainland)': 'China', 
    'Republic of Korea': 'South Korea', 
    # '': 'Falkland Islands', 
    'Syrian Arab Republic': 'Syria', 
    'Iran, Islamic Republic of': 'Iran',
    'China, Hong Kong SAR': 'Hong Kong', 
    "Democratic People's Republic of Korea" : 'North Korea', 
    'Bolivia (Plurinational State of)': 'Bolivia', 
    "Lao People's Democratic Republic": 'Laos',
    'Sao Tome and Principe': 'São Tomé and Príncipe',
    #'': 'Guernsey',
    # '': 'Gibraltar', No clue
    'Micronesia': 'Federated States of Micronesia', 
    # '': 'Faroe Islands', Cuonts towards denmark
    'Republic of North Macedonia': 'Macedonia',
    #'': 'Niue', Should count towards new zealand
    'United Republic of Tanzania: Mainland': 'Tanzania' # could also include zanzibar
    }
gdp_all['Country/Area'].replace(replace_dict, inplace=True)
```

```python
gdp_all.to_csv('gdp/gdp_all.csv', index=False)
```
